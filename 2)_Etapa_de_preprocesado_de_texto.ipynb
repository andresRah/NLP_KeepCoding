Etapa de preprocesado de texto

El alumno preparará una etapa de preprocesado de reviews que permita adecuar el formato de las mismas a uno más adecuado. Será la etapa previa al entrenamiento del modelo de sentimiento.

Todo el preprocesado deberá incluirse en una función de Python que contenga todo el procesado de texto. Esta función puede (es recomendable) contener otras funciones que realicen tareas más concretas (eliminar stopwords, eliminar signos de puntuación, etc.).


```python
# Importar Librerias Necesarias
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import unicodedata
```

```python
# Descargar stopwords y wordnet
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Cargar stopwords en inglés y lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Función para eliminar tildes
def eliminar_tildes(text):
    return ''.join((c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn'))
```

```python
# Función para eliminar símbolos y caracteres que no sean letras por medio de una regular expression
def eliminar_simbolos(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
# Función para eliminar stopwords
def eliminar_stopwords(text):
    return ' '.join([word for word in text.split() if word.lower() not in stop_words])
```

```python
# Función para lematización
def lematizacion(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
# Función de preprocesado de texto
def preprocesado_texto(df, text_column):
    # Eliminar valores nulos
    df = df.dropna(subset=[text_column])

    # Aplicar funciones de preprocesado
    df[text_column] = df[text_column].apply(eliminar_tildes)
    df[text_column] = df[text_column].apply(eliminar_simbolos)
    df[text_column] = df[text_column].apply(lambda x: x.lower()) # minusculas
    df[text_column] = df[text_column].apply(eliminar_stopwords)
    df[text_column] = df[text_column].apply(lematizacion)

    return df
```

```python
# Cargar archivo resultante del Punto 1. Análisis Exploratorio
file_path = 'corpus_one.csv'

df = pd.read_csv(file_path)

df.head(10)
```

```python
# Aplicar el preprocesado
df_preprocesado = preprocesado_texto(df, 'reviewText')
df_preprocesado.head(10)
```

```python
# Comparativo
print("Original:")
print(df['reviewText'][0])
print("Preprocesado:")
print(df_preprocesado['reviewText'][0])
```

```python
# guardar dataframe resultante de paso #2
df_preprocesado.to_csv('corpus_two.csv', index=False)
```
